```python
# Load the dataset
import pandas as pd
df = pd.read_csv("/content/titanic (1).csv")
df.head()
```

```python
print(df.columns)
```

```python
df.isnull().sum()

```

```python
# Fill missing pclass with median
if 'Pclass' in df.columns:  # Check if 'Age' column exists before proceeding

    print("Before fillna:")
    print(df['Pclass'].isnull().sum())
    df['Pclass'].fillna(df['Pclass'].median(), inplace=True)
    print("\nAfter fillna:")
    print(df['Pclass'].isnull().sum())
    print(df['Pclass'].dtype)
else:
    print("Column 'Pclass' not found in the DataFrame, skipping fillna.")
```

```python
if 'Embarked_Q' in df.columns:# Fill Embarked with mode
    df['Embarked_Q'].fillna(df['Embarked_S'].mode()[0], inplace=True)
else:
    print("Column 'Embarked_Q' not found in the DataFrame.")
    print("Column 'Embarked_S' not found in the DataFrame.")
```

```python
# Check if 'Parch' column exists before dropping
if 's' in df.columns:
    df.drop('Parch', axis=1, inplace=True)
else:
    print("Column 'Parch' not found in the DataFrame, skipping drop.")
```

```python
#categorical variables
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['Embarked_Q'], drop_first=True)
```

```python
#save the cleaned dataset
df.to_csv("titanic_cleaned.csv", index=False)
```

```python
#reload the dataset
df = pd.read_csv("titanic_cleaned.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Passenger Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Pclass', y='Fare', hue='Survived', palette='coolwarm', alpha=0.7)
plt.title('Pclass vs Fare (Colored by Survival)')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.show()
```

```python
sns.histplot(df['Fare'], bins=40, kde=True)
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
corr = df.corr(numeric_only=True)
# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```
